```python
#Importar librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import plotly.express as px
from matplotlib.animation import FuncAnimation
```

# Set de datos 1
Consta el set de datos del numero totales de nacimientos y defunciones por año

```python
dataset1= pd.read_csv('setdedatos.csv')
```

```python
#Crear el gráfico de líneas
plt.figure(figsize=(12, 6))
plt.plot(dataset1['año'], dataset1['Nacimientos'], label='Nacimientos', marker='o')
plt.plot(dataset1['año'], dataset1['Defunciones'], label='Defunciones', marker='o')

#Configurar el gráfico
plt.title('Nacimientos y Defunciones por Año')
plt.xlabel('Año')
plt.ylabel('Cantidad')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

#Mostrar el gráfico
plt.show()
```

# Set de datos 2
Set de datos de nacimientos y defunciones diferenciados por sexo en un periodo de 2023 hasta 2015

```python
#Cargar el primer dataset el cual contiene los nacimiento y defunciones diferenciados por sexo
dataset2= pd.read_csv('Dataset_nacimiento-defuncion_por_sexo.csv')
```

```python
dataset2.info()
```

```python
#Muestre general del dataset 1
dataset2
```

```python
#Grafico de distribucion de Nacimiento de ambos sexos mostrando asi las medidas estadisticas de nacimiento en ciertos años
px.box(dataset2, x='Año', y=['Nacimiento (Hombre)','Nacimiento (Mujer)'],
       title='Distribución de Nacimientos por Año')
```

```python
#Grafico de distribucion de Defunciones de ambos sexos mostrando asi las medidas estadisticas de nacimiento en ciertos años
px.box(dataset2, x='Año', y=['Defuncion(Hombre)','Defuncion (Mujer)'],
       title='Distribución de Defunciones por Año')
```

```python
#Grafico para mostrar todas las columnas relacionadas a nacimientos y defunciones tanto de hombre y mujeres en el paso del tiempo
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(go.Scatter(x=dataset2['Año'], y=dataset2['Nacimiento (Hombre)'],
                         name='Nacimiento(H)', hoverinfo='x+y'))

fig.add_trace(go.Scatter(x=dataset2['Año'], y=dataset2['Nacimiento (Mujer)'],
                         name='Nacimiento(M)', hoverinfo='x+y'))

fig.add_trace(go.Scatter(x=dataset2['Año'], y=dataset2['Defuncion(Hombre)'],
                         name='Defuncion(H)', hoverinfo='x+y'))

fig.add_trace(go.Scatter(x=dataset2['Año'], y=dataset2['Defuncion (Mujer)'],
                         name='Defuncion(M)', hoverinfo='x+y'))

fig.update_layout(title='Nacimientos y Defunciones por Año',
                  xaxis_title='Año',
                  yaxis_title='Cantidad')

fig.show()
```

# Set de datos 3
Consta de los nacimientos en ciertos años diferenciados por el rango de la madre

```python
#Cargar el segundo dataset el cual muestra el nacimiento diferenciado por el rango de edad de la madre
dataset3= pd.read_csv('Nacimiento_rango_edad_madre.csv')
```

```python
dataset3.info()
```

```python
#Muestreo general del dataset 2
dataset3
```

```python
#Grafica para mostrar la cantidad de nacimiento diferenciado por el rango de edad de la madre
import plotly.express as px

fig = px.line(dataset3, x='Año', y=dataset3.columns[1:],
              title='Cantidad por Año',
              labels={'value': 'Cantidad', 'variable': 'Rango de Edad'},
              hover_data={'Año': True})

fig.show()
```
